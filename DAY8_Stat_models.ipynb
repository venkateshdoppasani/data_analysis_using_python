```python
import numpy as np
from scipy import stats
```

```python
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
t_stat, p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python

print("T-statistic:",t_stat)
print("P-value:",p_value)
```

```python
#set significance values
alpha=0.05
```

```python
if p_value<alpha:
  print("Reject the null hypothesis. The means are significantly ")
else:
  print("Fail to reject null hypothesis. The means are not significantly")
```

```python
a=df['Efficiency'].mean()
b=df['Rec_Rate'].mean()
print("Mean of Efficiency :",a)
print("Mean of Recovery rate :",b)
```

```python
Group_A=np.array(df['Efficiency'])
Group_B=np.array(df['Rec_Rate'])
```

```python
t_stat,p_value =stats.ttest_ind(Group_A,Group_B,equal_var=False)
```

```python
print(t_stat)
print(p_value)
```

```python
if p_value < alpha:
  print("Reject the NULL hypothesis.The means are significant")
else:
  print("Failed to reject the NULL hypothesis.The means are different")
```
