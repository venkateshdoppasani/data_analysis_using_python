```python
#there are 5 subs psychology nanotechnology medicine onchology anthropology.2 two students have applied to do masters in 5 fields with a line plot show in which filed they can excel ,the cutoff for each subject is 70
# std1=[67,45,73,71,57]
#std2=[43,67,68,53,85]
sub=["psy","nano","medc","onc","anthro"]

std1=[67,45,73,71,57]
std2=[43,67,68,53,85]

cut=[70,70,70,70,70]

plt.scatter(sub,std1,label="student1",color='b',marker='*',linewidth=1)
plt.scatter(sub,std2,label="student2",linewidth=3)
plt.plot(cut,label="cutoff")

plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#ind vs autrs match is on live .ind has given 298 as target for aus,the runs for every 5 overs as follows:
#ind=[25,51,84,131,160,189,220,250,267,297]
#aus completed 25 overs
#aus=[15,41,94,110,151]
#plot overs vs runs for the countries ,ind with blue and yellow line,give appropriate labels and legend.the marker for ind is flag and for aus is plus.Take overs from arange functions .create runs as arrays,line width is 2
from matplotlib import pyplot as plt
import numpy as np
over1=np.arange(5,51,5)
ind=np.array([25,51,84,131,160,189,220,250,267,297])
ar=np.array([15,41,94,110,151])
over2=np.arange(5,26,5)
plt.plot(over1,ind,label="India",color='b',marker=">",linewidth=2)
plt.plot(over2,ar,label="Australia",color='y',marker="+",linewidth=2)
plt.legend()
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.grid()
plt.legend()
```
