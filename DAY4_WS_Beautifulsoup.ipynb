EXTRACTING HTML CODE OF A WEBPAGE
Program flow
1.Import libraries
2.save the url
3.Using requests.get(),access the ewb page
4.Using beautifulsoup(),accesss the HTML code

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests as rq
```

```python
url="https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox"
```

```python
page=rq.get(url)
```

```python
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
t=code.find("title")
print(t)
```

```python
print(t.text)
```

EXTRACTING A TABLE FROM URL

```python
from bs4 import BeautifulSoup
import requests as rq
```

```python
and1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
tables=rq.get(and1)
```

```python
code=BeautifulSoup(tables.text,'html')
print(code)
```

```python
b=code.find('table')
print(b)
```

```python
head=b.find_all('th')
print(head)
```

```python
headings=[i.text for i in head]
print(headings)
```

```python
import pandas as pd
mydf=pd.DataFrame(columns=headings)
print(mydf)
```

```python
rows=b.find_all('td')
print(rows)
```

```python
data=[i.text for i in rows]
print(data)
```

```python
st=0
stop=4
ind=0
for i in data:
  if stop<=40:
    mydf.loc[ind]=data[st:stop]
    st+=4
    stop+=4
    ind+=1
mydf
```

```python
mydf.to_csv("mydf.csv")
```
